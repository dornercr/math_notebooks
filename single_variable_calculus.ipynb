# Derivatives Study Guide

## Overview

This unit introduces derivatives, their geometric and physical interpretations, their importance in real-world applications, and how to compute them using fundamental calculus concepts.

---

## I. What is a Derivative?

### A. Definitions and Interpretations

- **Geometric Interpretation**: Slope of the tangent line to a graph at a point  
- **Physical Interpretation**: Instantaneous rate of change (e.g., velocity)  
- **Real-World Importance**: Used in science, engineering, economics, political science, etc.

### B. Key Idea

- The derivative at a point is:

$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$

---

## II. Geometric Interpretation: Tangent Line

### A. Tangent vs. Secant Line

- **Tangent line**: touches the graph at one point without crossing it  
- **Secant line**: passes through two points on the graph  

- Slope of the secant line:

$$
m_{\text{secant}} = \frac{f(x + h) - f(x)}{h}
$$

### B. Tangent Slope as a Limit

- As \( h \to 0 \), the secant becomes the tangent  
- Slope of tangent line:

$$
m_{\text{tangent}} = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

---

## III. Derivative Calculation

### A. Using Difference Quotient

$$
\frac{f(x + h) - f(x)}{h}
$$

- Divide by \( h \) and simplify

### B. Interpretation

- **Negative slope**: decreasing function  
- As \( h \to 0 \): slope magnitude increases or decreases depending on behavior

---

## IV. Applied Word Problem: Area Under a Tangent Line

### A. Problem

Find area of triangle formed by:
- Axes
- Tangent line to \( f(x) \)

### B. Tangent Line Equation

At point \( (x_0, f(x_0)) \):

$$
y - f(x_0) = f'(x_0)(x - x_0)
$$

### C. Solving for Intercepts

- **x-intercept**: Set \( y = 0 \) and solve for \( x \)  
- **y-intercept**: By symmetry, computed from equation

### D. Area Calculation

- Triangle Area:

$$
A = \frac{1}{2} \cdot x \cdot y
$$

- Constant area regardless of \( x \) — specific to \( f(x) \) form

## V. Notation for Derivatives

### A. Common Notations

- \( f'(x) \) (Newton's notation)  
- \( \frac{dy}{dx} \) (Leibniz notation)

### B. Interpretation

- All mean the same: the derivative of a function  
- Be aware of context when variables are reused (e.g., \( y \) as function and coordinate)
